```python
from time import sleep

data= []
bad_url = []
for page in pages:
    r = requests.get(page)
    soup = BeautifulSoup(r.content, 'lxml')
    
    tags = soup.find_all(class_='clearfix')
    row = tags[1]
    
    
    d = dict()
    try:
        d['prod_name'] = row.find("span",itemprop="name").text
    except:
        bad_url.append(page + str('name'))
        continue
    try:
        d['rating'] = row.find("span",class_="stars").get("title")
    except:
        bad_url.append(page + str('rating'))
        continue
    try:
        d['rating_count'] = tags[1].find("span",class_="prod-rating-copy").text
    except:
        bad_url.append(page + str('rtg_ct'))
        d['rating_count'] = None
    d['desc'] = tags[1].find("div", class_="p").text
    try:
        d['profile'] = row.find("div", class_ = "strengthGauge").get("title")
    except:
        bad_url.append(page + str('strength'))
        continue
    prod_size = row.find_all("a", {"class": "link linkToGrid"})
    
    sizes =[]
    for prod in prod_size:
        ea_size = prod.text
        sizes.append(ea_size)
    d['vitolas'] = sizes
    
    tables = row.find_all('table')
    df = pd.read_html(str(tables[0]))
    dfs = df[0]
    summary = {k: v.iloc[0, 1] for k, v in dfs.groupby(0)}
    d.update(summary)
#     d['details'] = pd.read_html(str(tables[0])))
    
#     details = row.find_all("span", class_="nobreak")
#     d['origin'] = details[1].text
#     d['flavored'] = details[2].text
#     d['pressed'] = details[3].text
#     d['has_tip'] = details[4].text
#     d['binder'] = details[5].text
#     d['filler'] = details[6].text
#     d['sweet'] = details[7].text
#     keys = ['wrapper', 'origin', 'flavored', 'pressed','has tip', 'binder', 'filler', 'sweet']
#     ind_tags = []
    
#     for key in keys:
        
#         formatted = str('<span>')+key.title()+str('</span>')
#         if formatted in str(tags[1]):
#             ind_tags.append(key)
#     d['details'] = ind_tags 


#     reviews = row.find_all("div",class_="prod-review")
#     top_reviews = reviews[0:50]

#     rev_list = []
    
#     for rev in top_reviews:
#         rev_dict = dict()
#         rev_dict['stars'] = rev.find("span", class_="stars").text
#         rev_dict['review_title'] = rev.find("div", class_="prod-review-title").text
#         rev_dict['review_body'] = rev.find("div", class_="prod-review-content").text
#         rev_list.append(rev_dict)
#     d['recent_rev'] = rev_list
    
    data.append(d)
    if len(data) %100 == 0:
        pd.DataFrame(data).to_csv('batch{}'.format(len(data)),index=False)
    sleep(randint(1,5))
```
