```python
cigar_df_feature_matrix = csr_matrix(df_final2.values)
```

```python
knn_search = NearestNeighbors(metric='cosine', algorithm='brute')
```

```python
knn_search.fit(cigar_df_feature_matrix)
```

```python
import autoFill as af
from ipywidgets import HTML, HBox

def open(value):
    show.value=value.new
    
options = list(df_final2.index)
autofill = af.autoFill(options,callback=open)

show = HTML('Result will be displayed here!')

display(HBox([autofill,show]))
```

```python
def get_key(val): 
    for key, value in dict(enumerate(options)).items(): 
         if val == value: 
            return key 
  

query_index = get_key(show.value)
distances, indices = knn_search.kneighbors(df_final2.iloc[query_index,:].values.reshape(1,-1), n_neighbors=11)

for i in range(0,len(distances.flatten())):

    if i == 0:
        print('Top Cigar Recommendations for: {}'.format(df_final2.index[query_index]))
    else:
        print('{}: {} with a Distance Score of: {}'.format(i, df_final2.index[indices.flatten()[i]],round(distances.flatten()[i],4)))
```
