```python
# Fill rating_count and rating null values with 0.

df_prod['rating_count'].fillna(0, inplace=True)
df_prod['rating'].fillna(0, inplace=True)
df_prod1 = df_prod[['prod_name', 'rating', 'rating_count']]
df_prod1.head()
```

```python
# Retrieve df_vitola

%store -r df_vitola
df_vitola.head()
```

```python
# Retrieve imputed data

df_itrans = pd.read_pickle('../data/working_data/impute.pkl')
```

```python
pd.DataFrame(df_itrans['Wrapper'].unique())
```

```python
# Transform common wrapper names to be uniform

df_itrans['Wrapper'] = df_itrans['Wrapper'].str.replace('San Andres', 'san_andres')
df_itrans['Wrapper'] = df_itrans['Wrapper'].str.replace('Sun Grown', 'sungrown')
df_itrans['Wrapper'] = df_itrans['Wrapper'].str.replace('Connecticut Broadleaf', 'connecticut_broadleaf')
df_itrans['Wrapper'] = df_itrans['Wrapper'].str.replace('Kentucky Fire-Cured', 'kentucky_fire_cured')
df_itrans['Wrapper'] = df_itrans['Wrapper'].str.replace('Costa Rican', 'costa_rican')
df_itrans['Wrapper'] = df_itrans['Wrapper'].str.replace('Pennsylvania Broadleaf', 'pennsylvania_broadleaf')
```

```python
df_itrans['Wrapper'].head()
```

```python
df_UW = df_itrans['Wrapper']
```

```python
# Display unique wrapper names

df_UW.unique()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Vectorize wrapper names

corpus = df_UW
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print((vectorizer.get_feature_names()))

CV_df = pd.DataFrame(X)
df_cvect = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
```

```python
print(len((vectorizer.get_feature_names())))
```

```python
df_cvect.head()
```

```python
# Scale vectorized wrapper dataframe

from sklearn.preprocessing import MinMaxScaler
data = df_cvect
scaler = MinMaxScaler()
scaler.fit(data)

X = scaler.transform(data)
print(scaler.data_max_)
df_wrapper = pd.DataFrame(X,columns=data.columns)
```

```python
# Add prefix to identify wrapper names in dataframe

df_wrapper = df_wrapper.add_prefix('w_')
df_wrapper.head()
```

```python
pd.DataFrame(df_itrans['Filler'].unique())
```

```python
# Transform common filler names to be uniform

df_itrans['Filler'] = df_itrans['Filler'].str.replace('United States', 'united_states')
df_itrans['Filler'] = df_itrans['Filler'].str.replace('Kentucky Fire-Cured', 'kentucky_fire_cured')
df_itrans['Filler'] = df_itrans['Filler'].str.replace('Costa Rican', 'costa_rican')
```

```python
df_itrans['Filler'].head()
```

```python
# Count vectorize filler names

corpus = df_UF
vectorizer = CountVectorizer()
Z = vectorizer.fit_transform(corpus)
print((vectorizer.get_feature_names()))

CV_df_f = pd.DataFrame(Z)
df_cvect_f = pd.DataFrame(Z.toarray(), columns=vectorizer.get_feature_names())
```

```python
print(len((vectorizer.get_feature_names())))
```

```python
df_cvect_f.head()
```

```python
# Scale vectorized filler dataframe

from sklearn.preprocessing import MinMaxScaler
data = df_cvect_f
scaler = MinMaxScaler()
scaler.fit(data)

Z = scaler.transform(data)
print(scaler.data_max_)
df_filler = pd.DataFrame(Z,columns=data.columns)
```

```python
# Add prefix to identify filler names

df_filler = df_filler.add_prefix('f_')
df_filler.head()
```

```python
# Trnsform column values to make them uniform

df_model_v['Has Tip'] = df_model_v['Has Tip'].str.replace('False', 'No')
df_model_v['Pressed'] = df_model_v['Pressed'].str.replace('True', 'Yes')
df_model_v['Pressed'] = df_model_v['Pressed'].str.replace('False', 'No')
df_model_v['Sweet'] = df_model_v['Sweet'].str.replace('True', 'Yes')
df_model_v['Sweet'] = df_model_v['Sweet'].str.replace('False', 'No')
```
