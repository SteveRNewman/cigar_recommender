```python
from time import sleep
review_body = []
prod_name = []
prod_names = []
data= []
rev_list = []
for page in pages:
    r = requests.get(page)
    soup = BeautifulSoup(r.content, 'lxml')
    
    tags = soup.find('h1')
    
    d = dict()
    d['prod_names'] = tags.text.strip()
    data.append(d)
    
    print(len(data))
    print(page)
    
    reviews = soup.find_all("div",class_="body")
    
    
    for rev in reviews:
        rev_dict = dict()
        rev_dict['prod_name'] = tags.text.strip()
        rev_dict['review_body'] = rev.text.strip()
        rev_list.append(rev_dict)

if len(rev_list) %500 == 0:
        pd.DataFrame(rev_list).to_csv('whisky_review{}.csv'.format(len(data)),index=False)

sleep(randint(1,5))
```
